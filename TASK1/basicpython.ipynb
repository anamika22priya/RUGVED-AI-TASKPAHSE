```python
#Q.3:
#A hill number must have digits that increase first,reach one peak and then decreases

num=input("Enter a number")#Take number input as a string
inc=False
dec=False
for i in range(len(num)-1):
    if num[i] < num[i+1] and not dec: #Digits increasing and decrease hasnt happend already
        inc=True
    elif num[i]>num[i+1] and inc: #Digits decreasing and increase has happend before
        dec=True
    else:
        print("Not a Hill Number")
        break
if inc and dec:
    print(f"{num} is a Hill Number")
else:
    print(f"{num} is not a Hill Number")
```

```python
#Q.4
def selection_sort_string(s):
    s = list(s) #convert to list because strings are immutable
    n = len(s)

    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if s[j] < s[min_index]:
                min_index = j
        s[i], s[min_index] = s[min_index], s[i] #swapping the characters. 

    return "".join(s) #remove whitespace


user_input=input("Enter a string to sort")
print(selection_sort_string(user_input))
```

```python
#Q.5:
def factorial(n):
    if n==0 or n==1:
        return n
    else:
        return n * factorial(n-1)
num=int(input("Enter the number"))
print(f"The factorial of {num} is {factorial(num)}")
```

```python
#Q.6:
def check_anagram(s1,s2):
    s1,s2=s1.lower().replace(" ",""), s2.lower().replace(" ","")
    if(len(s1)!=len(s2)):
        return False
    elif(sorted(s1)==sorted(s2)): #in-built sort function (uses TimSort)
        return True
    else:
        return False
    
input1=input("Enter the first string")
input2=input("Enter the second string")
if check_anagram(input1,input2):
    print(f"{input1} and {input2} are an anagram")
else:
    print("Not an anagram")
    

    
        
        

    
```

```python
#Q.7:
def fib(n):
    if n<=1:
        return n
    else:
        return fib(n-1)+fib(n-2)

num=int(input("Enter the number of terms")) 
for i in range(num):
    print(fib(i),end=" ")

```

```python
#Q.8:
def divide_string(s, n):
    length = len(s)

    # Check if division is possible
    if length % n != 0:
        print("Error: string length is not divisible by n")
        return

    parts = [s[i:i+n] for i in range(0, length, n)]

    # Check if all parts are the same
    if all(part == parts[0] for part in parts):
        print("Divided parts:", parts)
    else:
        print("Error: sequence in parts is not the same")

# Example usage
string = "abcdabcdabcdabcd"
n = 4
divide_string(string, n)
```
