```python
matches['Venue_cleaned'] = matches['Venue'].replace(stadium_map)
```

```python
#Q.12: Average runs scored in all matches in all venues 
merged = deliveries.merge(matches[['ID','Venue_cleaned']], on='ID')

avg_runs_venue = merged.groupby('Venue_cleaned')['total_run'].mean()
avg_runs_venue #Gives average runs per ball 
```

```python
#Q.13: Umpires who umpired the most:
umpires = pd.concat([matches['Umpire1'], matches['Umpire2']])

max_umpire = umpires.value_counts().idxmax()
max_umpire_count = umpires.value_counts().max()

print("Umpire who umpired the most:", max_umpire, "with", max_umpire_count, "matches")
```

```python
#Q.14: Total number of matches played in each season 
matches_per_season = matches.groupby('Season')['ID'].count()
matches_per_season
```

```python
#Q.15: Total runs scored in each season 
season_runs = deliveries.merge(matches[['ID','Season']], on='ID')
total_runs_season = season_runs.groupby('Season')['total_run'].sum()
total_runs_season
```

```python
#Q.16: Total runs scored by each batsman and display top 10
top_batsmen = deliveries.groupby('batter')['batsman_run'].sum().sort_values(ascending=False)
top_batsmen.head(10)

```

```python
#Q.17: total number of wickets taken by each bowler 
wickets = deliveries[deliveries['isWicketDelivery'] == 1]

wickets_per_bowler = wickets.groupby('bowler').size().sort_values(ascending=False)
wickets_per_bowler
```

```python
#Q.18: Batting averages and display top 10
runs = deliveries.groupby('batter')['batsman_run'].sum()

outs = deliveries[deliveries['isWicketDelivery']==1].groupby('player_out').size()

bat_avg = (runs / outs).dropna().sort_values(ascending=False)
bat_avg.head(10)
```

```python
#Q.19: Visualize toss decisions across seasons 
sns.countplot(data=matches, x='TossDecision')
plt.title('Toss Decisions Across All Seasons')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

```python
#Q.20:Total matches Vs Winning matches Vs Win rate for all teams  
total_matches = pd.concat([matches['Team1'], matches['Team2']]).value_counts()


wins = matches['WinningTeam'].value_counts()

win_rate = (wins / total_matches * 100).fillna(0)

team_stats = pd.DataFrame({'Total Matches': total_matches, 'Wins': wins, 'Win Rate (%)': win_rate})

team_stats.plot(kind='bar', figsize=(12,6))
plt.title('Total Matches vs Wins vs Win Rate')
plt.ylabel('Count / %')
plt.show()
```

```python
#Q.21: Distribution of the teams that won the matches 
sns.countplot(data=matches, x='WinningTeam')
plt.xticks(rotation=90)
plt.title('Distribution of Teams Who Won Matches')
plt.show()

```

```python
#Q.22: Visualize toss outcomes of all teams
sns.countplot(data=matches, x='TossWinner')
plt.xticks(rotation=90)
plt.title('Toss Outcomes of Teams')
plt.show()
```
